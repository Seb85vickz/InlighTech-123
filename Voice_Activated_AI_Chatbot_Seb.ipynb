```python
import speech_recognition as sr
import pyttsx3
import wikipediaapi
import webbrowser
import datetime
import time
import os
import subprocess
```

```python
# Text-to-speech
def speak(text: str) -> None:
    """
    Speak the given text aloud by printing it to the console (Colab-friendly fallback).
    """
    print(f"Bot: {text}")
```

```python
# Greeting based on time of day
def wishMe() -> None:
    """
    Greet the user based on the current time.
    """
    hour = int(datetime.datetime.now().hour)

    if 0 <= hour < 12:
        greet = "Good morning!"
    elif 12 <= hour < 18:
        greet = "Good afternoon!"
    else:
        greet = "Good evening!"

    speak(greet)
    speak("I am your voice-activated chatbot. How can I help you today?")
```

```python
# Listen from microphone and return recognized string
def takeCommand() -> str:
    """
    Accepts text input from the user instead of microphone input (Colab-friendly fallback).
    Returns the input text in lowercase.
    """
    query = input("User: ")
    return query.lower()
```

```python
# Wikipedia helper
wiki = wikipediaapi.Wikipedia(
    user_agent="VoiceChatbot/1.0 (student project)",
    language="en"
)

def search_wikipedia(topic: str) -> None:
    """
    Fetch a short summary from Wikipedia for the given topic and speak it.
    """
    page = wiki.page(topic)
    if not page.exists():
        speak("I could not find anything on Wikipedia for that topic.")
        return

    summary = page.summary[:500]  # first 500 characters
    speak("According to Wikipedia")
    speak(summary)
```

```python
# 1. Open Google
def main() -> None:
    wishMe()

    while True:
        query = takeCommand()
        if query == "none":
            continue

        if "open google" in query:
            speak("Opening Google.")
            webbrowser.open("https://www.google.com")

        # 2. Open YouTube
        elif "open youtube" in query:
            speak("Opening YouTube.")
            webbrowser.open("https://www.youtube.com")

        # 3. Wikipedia search
        elif "wikipedia" in query or "search wikipedia" in query:
            speak("What should I search on Wikipedia?")
            topic = takeCommand()
            if topic != "none":
                search_wikipedia(topic)

        # 4. Tell the time
        elif "time" in query or "what is the time" in query:
            strTime = datetime.datetime.now().strftime("%H:%M")
            speak(f"The time is {strTime}")

        # 5. Greeting / small talk
        elif "hello" in query or "hi" in query:
            speak("Hello! How are you?")
        elif "who are you" in query:
            speak("I am your voice activated AI chatbot, created using Python.")
        elif "how are you" in query:
            speak("I am just a program, but I am running fine. Thank you for asking.")

        # 6. Custom commands (notes and to-do)
        elif "write a note" in query or "make a note" in query:
            speak("What should I write?")
            note = takeCommand()
            if note != "none":
                add_note(note)

        elif "add task" in query or "add to do" in query or "add todo" in query:
            speak("What task should I add?")
            task = takeCommand()
            if task != "none":
                add_todo(task)

        # 7. System commands (comment out if you don't want real shutdown)
        elif "shutdown" in query and "system" in query:
            speak("Shutting down the system. Please make sure all your work is saved.")
            # os.system("shutdown /s /t 1")  # Windows example

        elif "restart" in query and "system" in query:
            speak("Restarting the system.")
            # os.system("shutdown /r /t 1")  # Windows example

        # 8. Exit
        elif "quit" in query or "exit" in query or "stop" in query:
            speak("Goodbye!")
            break

        else:
            speak("I am not sure how to handle that command yet.")
            print("Unrecognized command:", query)


if __name__ == "__main__":
    main()
```
